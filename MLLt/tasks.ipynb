```python
def matrix_dot_vector(a:list[list[int|float]],b:list[int|float])-> list[int|float]:
    c = [0] * len(b)
    for i in range(len(a)):
        if(len(a[i]) != len(b)):
            return -1
        for j in range(len(a[i])):
            c[i] += a[i][j]*b[j]
    return c
```

```python
matrix_dot_vector(a = [[1,2],[2,4]], b = [1,2])
```

```python
def calculate_covariance_matrix(vectors: list[list[float]]) -> list[list[float]]:
    covariance_matrix = []
    mean_dif = []
    for vector in vectors:
        temp = []
        mean = sum(vector)/len(vector)
        for item in vector:
            temp.append(item - mean)
        mean_dif.append(temp)
            
    for a in mean_dif:
        temp = []
        for b in mean_dif:
            mult = 0
            for i in range(len(a)):
                mult += a[i]*b[i]
            temp.append(mult/(len(a)-1))
        covariance_matrix.append(temp)   
    return covariance_matrix
```

```python
calculate_covariance_matrix([[1, 2, 3], [4, 5, 6]])
```

```python
import numpy as np
def solve_jacobi(A: np.ndarray, b: np.ndarray, n: int) -> list:
    x = [0]*len(A)
    for _ in range(n):
        t = [0] * len(A)
        t = []
        for i in range(len(A)):
            s = 0
            for j in range(len(A[i])):
                if(j != i):
                    s += A[i][j]*x[j]
            t.append(1/A[i][i]*(b[i] - s))
        x = t
    return x
```

```python
solve_jacobi(A = [[5, -2, 3], [-3, 9, 1], [2, -1, -7]], b = [-1, 2, 3], n=2)
```

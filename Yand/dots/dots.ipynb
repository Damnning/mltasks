```python
from random import uniform
from math import cos, sin, pi
import matplotlib.pyplot as plt
```

```python
def generate1():
    a = uniform(0, 1)
    b = uniform(0, 1)
    return (a * cos(2 * pi * b), a * sin(2 * pi * b))

def generate2():
    while True:
        x = uniform(-1, 1)
        y = uniform(-1, 1)
        if x ** 2 + y ** 2 > 1:
            continue
        return (x, y)
```

```python
def add_to_input(inp):
    with open('input.txt', 'w') as file:
    	file.write(inp)
        
def get_gen1_str():
    gen1 = ''
    for _ in range(1000):
        x, y = generate1()
        gen1 += (str(x) + ' ' + str(y) + ' ')
    return gen1 + '\n'

def get_gen2_str():
    gen2 = ''
    for _ in range(1000):
        x, y = generate2()
        gen2 += (str(x) + ' ' + str(y) + ' ')
    return gen2 + '\n'
```

```python
inp = get_gen1_str()
inp += get_gen1_str()
inp += get_gen2_str()
inp += get_gen1_str()
add_to_input(inp)
```

```python
def define_generator(gen):
    gen = gen[:-1]
    n = 0
    nums = list(map(float, gen.split(' ')))
    for i in range(1000):
        if(nums[i]**2 + nums[i+1]**2 < 0.5):
            n+=1
    return n
```

```python
def draw_points(gen1, gen2):
    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    plt.grid()
    gen1 = gen1[:-1]
    gen2 = gen2[:-1]
    nums1 = list(map(float, gen1.split(' ')))
    nums2 = list(map(float, gen2.split(' ')))
    for i in range(1000):
        plt.plot(nums1[i], nums1[i+1], marker="o", markersize=1, markeredgecolor="red", markerfacecolor="green")
        plt.plot(nums2[i], nums2[i+1], marker="o", markersize=1, markeredgecolor="blue", markerfacecolor="green")
    plt.show()
```

```python
def count_narrow(gen):
    n = 0
    nums = gen.split(' ')
    nums = list(filter(None, nums))
    nums = list(map(float, nums))
    for i in range(0, len(nums), 2):
        if(nums[i]**2 + nums[i+1]**2 < 0.5):
            n+=1
    return n
        
def add_to_output(res):
    with open('output.txt', 'w') as file:
    	file.write(f'{res}\n')

def read_input():
    with open('input.txt') as file:
    	inp = file.read()
    inp = inp.split('\n')
    return inp

inp = read_input()
result = ''
for i in inp:
    if(i != ''):
        c = count_narrow(i)
        if(c > 600):
            result += '1\n'
        else:
            result += '2\n'
add_to_output(result)
```
